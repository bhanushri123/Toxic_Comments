```python
from tensorflow.keras.layers import TextVectorization
```

```python
X = df['comment_text']
Y = df[df.columns[2:]].values
```

```python
df[df.columns[2:]].values
```

```python
X
```

```python
Y
```

```python
MAX_FEATURES = 20000
# number of words in the vocab
```

```python
vectorizer   = TextVectorization(max_tokens = MAX_FEATURES,output_sequence_length=1800,output_mode = 'int')
# output_sequence_length this is the max input length taht we rae taking from the comments
```

```python
X.values
```

```python
vectorizer.adapt(X.values)
```

```python
vectorizer.get_vocabulary()
# this gives us an array of vocabs where the position of each word tells us the number given top that word
```

```python
vectorizer("Hello my name, is Bhanu")[:5]
# working of vectorizer
```

```python
vectorized_text = vectorizer(X.values)
```

```python
vectorized_text
```

```python
# We will build a data pipeline now
# MCSHBAP map , cache,shuffle , batch , prefetch
```

```python
dataset  = tf.data.Dataset.from_tensor_slices((vectorized_text,Y))
dataset = dataset.cache()
dataset = dataset.shuffle(16000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8)
```

```python
batch_X,batch_Y = dataset.as_numpy_iterator().next()
```

```python
batch_Y.shape
```

```python
train = dataset.take(int(len(dataset)*.7))
val = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))
```

```python
train_generator = dataset.as_numpy_iterator()
```

```python
train_generator.next()
```

```python
from keras.layers import Dense,LSTM,Bidirectional,Embedding,Dropout
from keras.models import Sequential,Model
```

```python
model = Sequential()
model.add(Embedding(MAX_FEATURES+1, 32))
model.add(Bidirectional(LSTM(32, activation ='tanh')))
model.add(Dense(128, activation ='relu'))
model.add(Dense(256, activation ='relu'))
model.add(Dense(128, activation ='relu'))
model.add(Dense(6, activation ='sigmoid'))

```

```python
model.compile(loss = 'BinaryCrossentropy',optimizer = 'Adam')
```

```python
model.summary()
```

```python
history = model.fit(train, epochs = 1,validation_data = val)
```

```python
history.history
```
